### Tokenizer


```python
import re

### Compile the regex into a regex object ###
r = re.compile(r"ab+c")

### Check whether the entire input matches ###
print(r.fullmatch("abc"))
print(r.match("abcd"))
# Prints an object describing the match

print(r.fullmatch("axc"))
# Prints "None"


### Search a string for matches ###
print(r.search("the abc book"))
# Prints "<re.Match object; span=(4, 7), match='abc'>"

print(r.findall("the abc book. abbbc!"))
# Prints "['abc', 'abbbc']"


### Check whether a pattern occurs at an index ###
print(r.match("the abc book", 4))
# Prints "<re.Match object; span=(4, 7), match='abc'>"

print(r.match("the abc book", 5))
# Prints "None", because it didn't match at index 5
```

```python
# Whitespace regex
white = re.compile("[\t\n ]+")

print(white.match("abcd"))
print(white.match("  abcd"))
print(white.match("\n \nabcd").end())
```

```python
# Integer regex
intr = re.compile("[0-9]+")

print(intr.match("93"))
print(intr.match("abc"))
print(intr.match("100028"))
print(intr.match("100028f").group())
```

```python
# Identifier regex
identr = re.compile("[a-zA-Z_][a-zA-Z_0-9]*")

print(identr.match("9abc"))
print(identr.match("abc9   "))
print(identr.match("_a9_23   "))
```

```python
import re

line = re.compile("(#|//)(.*)\n")
print(line.match("# bla bla \n 123"))
print(line.match("// bla bla \n 123"))
```

```python
import re

multi_line_comment = re.compile("(/\\*)(.|\n)*(\\*/)")
print(multi_line_comment.match("/* hey* */ ok"))
print(multi_line_comment.match("/* \n multi \n line\ncomment */ after"))
```
